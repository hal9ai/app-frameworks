```python
import io
import param
import panel as pn
import pandas as pd
import numpy as np
import tensorflow as tf
from PIL import Image

pn.extension()
```

```python
class BrightnessApp(param.Parameterized):
    data = param.DataFrame()
    
    file_input = param.Parameter()
    
    def __init__(self, **params):
        super().__init__(file_input=pn.widgets.FileInput(), **params)
        self.jpg_pane = pn.Row(pn.pane.JPG(height=400, sizing_mode="stretch_width"), pn.pane.JPG(height=400, sizing_mode="stretch_width"))
        self.model =  tf.keras.models.load_model('model_1.h5')  # load the pretrained mode
        
    @pn.depends("file_input.value", watch=True)
    def save_image(self):
        value = self.file_input.value
        if value:
            img_file = open('original.jpg', 'wb')                  # write the decoded data back to original format in  file
            img_file.write(value)
            img_file.close()
            self.jpg_pane[0] = pn.pane.JPG('original.jpg', width=300)
            
            img = np.array(Image.open('original.jpg'))
            fotos_mod = []

            img = img / 255.  # normalices the array
            fotos_mod.append(img)  # 'save' the image in fotos_mod

            x = np.array(fotos_mod)  # turns fotos_mod in an array

            test_predictions = self.model.predict(x)

            b = test_predictions.shape[1:4]
            c = test_predictions.reshape(b)
            modify = Image.fromarray((c * 255).astype(np.uint8))

            modify.save('modify.jpg')

            self.jpg_pane[1] = pn.pane.JPG('modify.jpg', width=300)

            self.data = np.asarray(Image.open('original.jpg'))
        else:
            print("error")


    def view(self):
        return pn.Column(
            "#Modify Brightness in Images",
            self.file_input,
            self.jpg_pane,
        )
    
brightness_app = BrightnessApp()

brightness_app_view = brightness_app.view()
brightness_app_view
```

```python
component = pn.Column(brightness_app_view)

component.show()
```
